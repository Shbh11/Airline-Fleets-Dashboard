```python
import pandas as pd
import random
```

```python
filename = "Original Fleet Data.csv"
```

```python
data = pd.read_csv(filename)
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
cols_to_clean = data.columns[3:]
#len(col_to_clean)
```

```python
# All missing(NaN) values will be filled with a random number between the min and max value for the respective column
```

```python
for i in range(3, 8):
    df=data.iloc[:,i]
    df.fillna(random.randrange(df.min(),df.max()), inplace=True)
```

```python
data['Average Age'].fillna(random.randrange(1,39), inplace=True)
```

```python
# Conveting object data type to float64 and also removing the $ symbol
data['Unit Cost']=data['Unit Cost'].apply(lambda x: str(x).replace('$', ''))
data['Unit Cost']=data['Unit Cost'].astype('float64')
data['Total Cost (Current)']=data['Total Cost (Current)'].apply(lambda x: str(x).replace('$', '').replace(",",""))
data['Total Cost (Current)']=data['Total Cost (Current)'].astype('float64')
```

```python
data['Unit Cost'].fillna(random.randrange(data['Unit Cost'].min(),data['Unit Cost'].max()), inplace=True)
data['Total Cost (Current)'].fillna(random.randrange(data['Total Cost (Current)'].min(),data['Total Cost (Current)'].max()), inplace=True)
```

```python
data.info()
```

```python
#Renaming columns
data['Unit Cost'].rename('Unit Cost (USD)')
data['Total Cost (Current)'].rename('Total Cost (USD)')
data['Average Age'].rename('Average Age (Years)')
```

```python
data['Aircraft Manufacturer'] = data['Aircraft Type'] # Creating new column 'Aircraft Manufacturer'
```

```python
data=data.iloc[:,1:] # Removing index column
```

```python
data.head()
```

```python
d = data["Aircraft Manufacturer"].str.rpartition(' ', expand=True) # Splitting the column based on last occurence of " " character
```

```python
d.head() 
```

```python
d=d.iloc[:,[0,2]] # We only need columns 0 and 2
d.head()
```

```python
data.reset_index(drop=True, inplace=True)
d.reset_index(drop=True, inplace=True)
```

```python
final = pd.concat([data, d], axis = 1) # Concatenating the 2 dataframes
final.head()
```

```python
final.rename(columns={0: 'Aircraft Manufacturer', 2: 'Aircraft Model'})
```
